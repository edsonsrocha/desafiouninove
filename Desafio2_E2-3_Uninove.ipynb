```python
# Importando bibliotecas necessárias para o desafio 2
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
```

- Caso aconteça algum erro ao carregar o "driver" relacionado à versão do Google Chrome, verificar a versão do navegador instalado e baixar o executável do "driver" referente ao mesmo. Segue o link para download: https://chromedriver.chromium.org/downloads

```python
# Criando variável 'driver' referente ao 'selenium' e abrindo o navegador para os procedimentos automatizados
driver = webdriver.Chrome('chromedriver.exe')

# Criando variável referente ao site
site_estatisticas = 'https://1xbet.whoscored.com/Statistics'

# Acessando o site na página de estatísticas
driver.get(site_estatisticas)
time.sleep(3)

# Selecionando o ícone referente ao campeonato brasileiro
icone_brasileirao = driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[3]/ul/li[9]/a')
icone_brasileirao.click()
time.sleep(3)

# Criando variáveis para obter os elementos da página
elemento = driver.find_element_by_xpath('/html/body/div[4]/div[3]/div[1]/div[9]/div[2]/div[2]')
resposta = elemento.get_attribute('innerHTML')
conteudo_site = BeautifulSoup(resposta, 'html.parser')

# Selecionando elementos da página para efetuar a raspagem dos dados da mesma
elementos_tab_th = conteudo_site.find_all('tr', class_="standings-20566-general-header") # Nomes das colunas
elementos_tab_th = elementos_tab_th[0].find_all('th') # Nomes das colunas
elementos_tab_td = conteudo_site.find_all('td') # Dados estatísticos

# Criando lista com os nomes da colunas
colunas_tabela = []
for coluna in elementos_tab_th:
    colunas_tabela.append(coluna.text)

# Criando lista de listas referente aos dados da tabela
dados_equipes = []

# Criando variáveis para funcionar dentro do 'Looping'
qtd_colunas = len(colunas_tabela)
idx = 0 # Índice referente ao primeiro elemento referente aos dados das equipes
# Quantidade de equipes na página (quantidade de dados de todos os times / quantidade de colunas da tabela)
qtd_equipes = int(len(elementos_tab_td) / len(elementos_tab_th))

# 'Looping' para armazenar os dados das equipes na lista 'dados_equipes'
for _ in range(qtd_equipes):
    equipe = []
    # 'Looping para buscar os dados de cada um dos equipes disponíveis na página'
    for _ in range(qtd_colunas):
        equipe.append(elementos_tab_td[idx].text)
        idx += 1
    dados_equipes.append(equipe)

# Criando 'DataFrame' da tabela do Campeonato Brasileiro Série A
tabela_brasileiro = pd.DataFrame(dados_equipes, columns=colunas_tabela)
tabela_brasileiro
```

```python
# Acessando novamente o site na página de estatísticas
driver = webdriver.Chrome('chromedriver.exe')
site_estatisticas = 'https://1xbet.whoscored.com/Statistics'
driver.get(site_estatisticas)
time.sleep(3)

# Criando a variável com o nome da equipe desejada
string_equipe = 'Ajax'

# Inserindo o nome da equipe no campo de pesquisa
texto_pesquisa = driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[3]/div/form/input')
texto_pesquisa.clear()
texto_pesquisa.send_keys(string_equipe)
time.sleep(1)

# Clicando no botão de pesquisa
btn_pesquisa = driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[3]/div/form/div/span')
btn_pesquisa.click()
time.sleep(3)

# Selecionando a primeira equipe da lista de equipes disponíveis no resultado da pesquisa
btn_primeira_equipe = driver.find_element_by_xpath('/html/body/div[4]/div[3]/div[1]/div[2]/table[1]/tbody/tr[2]/td[1]/a')
btn_primeira_equipe.click()
time.sleep(3)

# Selecionando a guia 'Offensive' na tabela "Ajax Squad"
guia_offensive= driver.find_element_by_xpath('/html/body/div[4]/div[6]/div/ul/li[3]/a')
guia_offensive.click()
time.sleep(3)

# Criando variáveis para obter os elementos da página
elemento = driver.find_element_by_xpath('/html/body/div[4]/div[6]/div/div[3]')
resposta = elemento.get_attribute('innerHTML')
conteudo_site = BeautifulSoup(resposta, 'html.parser')

# Selecionando elementos da página para efetuar a raspagem dos dados da página
elementos_tab_th = conteudo_site.find_all('th') # Nomes das colunas
elementos_tab_td = conteudo_site.find_all('td') # Dados estatísticos

# Criando lista das colunas da tabela 'Offensive'
colunas_offensive = []

# 'Looping' para armazenar os nomes das colunas da tabela
for elem in elementos_tab_th:
    colunas_offensive.append(elem.text)
# Renomeando a primeira coluna que será excluída futuramente (antigo nome = 'Player', o mesmo nome da segunda coluna), os dados delas são idênticos ao da segunda,
# mas com um algarismo numérico junto com o primeiro nome de cada jogador
colunas_offensive[0] = 'Remover'

# Criando lista de listas referente aos dados da tabela
dados_jogadores = []
qtd_colunas = len(colunas_offensive)
idx = 0 # Índice referente ao primeiro elemento referente aos dados dos jogadores 'offensive'

# Criando variável com a quantidade de jogadores desejados
qtd_jogadores = 5

# 'Looping' para armazenar os dados dos jogadores na lista 'dados_jogadores'
for _ in range(qtd_jogadores):
    jogador = []
    # 'Looping para buscar os dados de cada um dos jogadores disponíveis na página'
    for _ in range(qtd_colunas):
        jogador.append(elementos_tab_td[idx].text.strip())
        idx += 1
    dados_jogadores.append(jogador)

# Criando 'DataFrame' da tabela 'Ofensive'
tab_offensive = pd.DataFrame(dados_jogadores, columns=colunas_offensive)
tab_offensive.drop(columns='Remover')
```
